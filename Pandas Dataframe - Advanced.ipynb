## **Import Pandas & Load Dataset**

```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/insurance.csv")
data.head()
```

## **Membuat kolom BMI_category**

```python
data["BMI_Category"] = data['bmi'].apply(lambda x: "High BMI" if x>=50 else ("Medium BMI" if x>=30 and x<50 else "Low BMI"))
data.head()
```

## **Mengubah Data Point Kolom Sex, Region, dan BMI_Category Menjadi Uppercase**

```python
data[['sex', 'region', 'BMI_Category']] = data[['sex', 'region', 'BMI_Category']].apply(lambda x: x.astype(str).str.upper())
data.head()
```

## **Menentukan Nilai Minimum, Maximum, dan Mean dari BMI**

```python
BMI = pd.pivot_table(data, values = ['bmi'], index = 'BMI_Category', aggfunc = {'bmi':['min', 'max', 'mean']}).round(2)
BMI
```
