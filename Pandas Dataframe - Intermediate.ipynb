## **Import Pandas**

```python
import pandas as pd
```

## **Reading data from file data.csv and data_2.csv**

```python
data1 = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/data.csv")
print(data1.shape)
data1.head()
```

```python
data2 = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/data_2.csv")
print(data2.shape)
data2.head()
```

## **Full Outer Join data1 & data2**

```python
final = data1.merge(data2, how='outer', on='employee_id')
print(final.shape)
final
```

## **Show only rows which 'awards_won_y' is NULL**

```python
filtered_final = final.loc[(final['awards_won_y'].isnull())]
print(filtered_final.shape)
filtered_final
```

## **Pick columns which only exists in data.csv**

```python
filtered_final.iloc[:,1:14]
```

## **Rename the columns as the same as data.csv's column name**

```python
filtered_final_ren = filtered_final.rename({'department_x' : 'department', 'region_x' : 'region', 'education_x' : 'education', 'gender_x' : 'gender',
                                            'recruitment_channel_x' : 'recruitment_channel', 'no_of_trainings_x' : 'no_of_trainings', 'age_x' : 'age',
                                            'previous_year_rating_x' : 'previous_year_rating', 'length_of_service_x' : 'length_of_service', 'awards_won_x' : 'awards_won',
                                            'avg_training_score_x' : 'avg_training_score', 'is_promoted_x' : 'is_promoted'}, axis=1)
print(filtered_final_ren.shape)
filtered_final_ren.iloc[:,1:14]
```
