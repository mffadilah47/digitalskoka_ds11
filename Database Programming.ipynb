## **Joining table cb_stations & cb_trips**
filter tripduration > 86400sec (1 hari)

```python
sql = """
      select a.name, a.region_id, a.rental_methods, b.tripduration from batch_11.cb_stations as a JOIN batch_11.cb_trips as b on a.station_id = b.end_station_id where tripduration > 86400
      """
data = pd.read_sql_query(sql, conn)
data.head()
```

## **Creating Pivot Table with Tripduration Mean for Each Station**

```python
summary = pd.pivot_table(data, values = ['tripduration'], index = 'name', aggfunc = {'tripduration' : ['mean']})
summary.index.names = ['station name']
summary
```
